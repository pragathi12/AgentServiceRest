```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x_flen = 200.0
y_flen = 200.0
x_ppnt = 256.0
y_ppnt = 256.0
X,Y,Z=-50,20,200
x_a=((x_flen*(X))/Z)+x_ppnt
y_a=((y_flen*(Y))/Z)+y_ppnt
print(x_a)
print(y_a)
D_3_image = plt.figure()
pix = D_3_image.add_subplot(111, projection='3d')
pix.scatter(x_a, y_a, 200, c='r', marker='o', label='object')
pix.scatter(X, Y, Z, c='g', marker='s', label='Camera')
pix.legend()
```

# Question 3 Change in angle

```python
x_flen = 200.0
y_flen = 200.0
c=np.array([[0,0,200],
           [50,0,200],
           [50,50,200],
           [0,50,200],
           [0,0,250],
           [50,0,250],
           [50,50,250],
           [0,50,250]])
angle=np.array([[np.cos(np.radians(35)),0,np.sin(np.radians(35))],
                 [0,1,0],
                 [-np.sin(np.radians(35)),0,np.cos(np.radians(35))]])
transition=np.array([200,0,0])
rotate=(angle @ c.T).T+ transition
l1=[]
fig,ax = plt.subplots()

for i in rotate:
    x_axis=((x_flen*i[0])/Z)+x_ppnt
    y_axis=((y_flen*i[1])/Z)+y_ppnt
    ax.plot(x_axis, y_axis, marker='o', linestyle='-')
    l1.append((x_axis,y_axis))
ax.scatter(X, Y, c='b', marker='s', label='Camera')
print("list:",np.array(l1).astype(int))
print(rotate.astype(int))
x = [point[0] for point in l1]
y = [point[1] for point in l1]
for i in range(4):
        ax.plot([x[i], x[i + 4]], [y[i], y[i + 4]], linestyle='-', color='b')
        ax.plot([x[i], x[i]], [y[i+1], y[i +2]], linestyle='-', color='b')
        ax.plot([x[i+3], x[i+4]], [y[i+3], y[i + 4]], linestyle='-', color='b')








```

```python
c=np.array([[0,0,200],
           [50,0,200],
           [50,50,200],
           [0,50,200],
           [0,0,250],
           [50,0,250],
           [50,50,250],
           [0,50,250]])
x_flen = 200.0
y_flen = 200.0
x_ppnt = 500.0
y_ppnt = 500.0
skew =[0,10]
p1=[]
p2=[]
for i in c:
    X=i[0]
    Y=i[1]
    Z=i[2]
    for i in skew:
        if(i==0):
            x = ((x_flen * X+i*Y) / Z) + x_ppnt
            y = (y_flen * Y / Z) + y_ppnt
            p1.append((x,y))
        if(i==10):
            x = ((x_flen * X+i*Y) / Z) + x_ppnt
            y = (y_flen * Y / Z) + y_ppnt
            p2.append((x,y))
print('list:',p1)
ax.plot(p1)
print('list:',p2)
ax.plot(p2)
```
